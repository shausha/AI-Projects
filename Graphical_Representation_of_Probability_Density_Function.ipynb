```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
# Define the PDF function (normal distribution)
def pdf(x, mean, std_dev):
  return np.exp(-(x - mean)**2 / (2 * std_dev**2)) / (std_dev * np.sqrt(2 * np.pi))
# Given mean and standard deviation for the stock returns
mean_return = 0.001 # Mean daily return (0.1%)
std_dev_return = 0.012 # Standard deviation of daily returns (1.2%)
# Define the lower and upper bounds for the desired return range
lower_bound_return = float(input("Enter the lower bound:")) # -2%
upper_bound_return = float(input("Enter the upper bound:")) # 2%
# Calculate the probability under the curve between the bounds
probability, _ = integrate.quad(lambda x: pdf(x, mean_return, std_dev_return),
lower_bound_return, upper_bound_return)
print("Probability of return between bounds:", probability)

# Create x values for the plot
x_vals = np.linspace(mean_return - 4 * std_dev_return, mean_return + 4 *
std_dev_return, 1000)
# Create y values using the PDF function
y_vals = pdf(x_vals, mean_return, std_dev_return)
# Plot the PDF
plt.plot(x_vals, y_vals, label='PDF',color='pink')
# Shade the region under the curve between bounds
plt.fill_between(x_vals, y_vals, where=((x_vals >= lower_bound_return) & (x_vals <=
upper_bound_return)), color='black',alpha=0.7, label='Probability Density')
# Add vertical lines for bounds
plt.axvline(lower_bound_return, color='r', linestyle=':', label='Lower Bound')
plt.axvline(upper_bound_return, color='g', linestyle='--', label='Upper Bound')
plt.xlabel('Daily Return')
plt.ylabel('Probability Density')
plt.title('Probability Density of Daily Stock Returns P(-2%<X<2%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
