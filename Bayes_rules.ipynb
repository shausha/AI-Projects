```python
import matplotlib.pyplot as plt
# prior prob by user input
prevalence_condition_prior=int(input("The prevalence of the condition(%):"))
prevalence_cond=prevalence_condition_prior/100
print("Prior Probability P(A):",prevalence_cond)
#likelyhood
tes=int(input("The probability of test positive given having the condition(%):"))
test_positive_wi_cond=tes/100
print("Probabilty of test given condition present P(B|A):",test_positive_wi_cond)
#to calculate evidence
tes1=int(input("The probability of testing positive given not having the condition(%):"))
test_positive_wio_cond=tes1/100
print("Probability of test positive with no condition P(B|¬A):",test_positive_wio_cond)
not_cond=1-prevalence_cond
print("Probability of not having condition P(¬A):",not_cond)
#evidence
evi=test_positive_wi_cond*prevalence_cond+test_positive_wio_cond*not_cond
print("Evidence P(B):",evi)
#posterior prob
post_prob=test_positive_wi_cond * prevalence_cond/evi

print("Posterior probability P(A|B):",post_prob)
labels = ['Prior Probability', 'Posterior Probability']
values = [prevalence_cond, post_prob]
colors = ['black', 'purple']
plt.bar(labels, values, color=colors, label=['Prevalence condition', 'Test +ve|conditon'])
plt.ylabel('Probability')
plt.title('Probability of Disease prevalence before and after test')
plt.legend()
plt.show()
```
