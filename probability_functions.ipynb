## NORMAL DISTRIBUTION:

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
#setting a random variable
dat=stats.norm.rvs(size=400,loc=0,scale=10)
print('Random Numbers:',dat)
#plotting a histogram with colour
graph=sns.distplot(dat,bins=50,color='r')
#setting label for graph
graph.set(xlabel='marks',ylabel='average')
#setting legend for graph
plt.legend(['single element'])
plt.title('normal distribuition')
plt.show()
```

## BINOMIAL DISTRIBUTION:

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a random sample of binomially distributed data
x = random.binomial(n=10, p=0.5, size=30)
print('Random Numbers:', x)

# Create a histogram-like plot using seaborn.histplot() and add labels for the legend
sns.histplot(random.binomial(n=10, p=0.5, size=10000), bins=10, kde=False, color='y', label='n=10, p=0.5')

# Add labels and a title to the plot
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (Discrete)')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

## BERNOULLI’S DISTRIBUTION:

```python
import matplotlib.pyplot as plt # For creating plots
import seaborn as sns # For data visualization
from scipy.stats import bernoulli # For generating Bernoulli-distributed random numbers
data_bern = bernoulli.rvs(size=100, p=0.6)
print(data_bern)
# Create a histogram and plot the distribution of the generated data
ax = sns.distplot(data_bern, bins=2, kde=False, color='red', label='marks',
hist_kws={"linewidth": 10, 'alpha': 1})
# Setting labels, titles and legend
ax.set(xlabel='Bernoulli', ylabel='Frequency')
plt.title('Bernoulli Distribution (Discrete)')
plt.legend()
# Show the plot
plt.show()
```

## UNIFORM DISTRIBUTION:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Generate random data from a continuous uniform distribution
low,high,size = 0,5,10
data = np.random.uniform(low, high, size)
low1,high1,size1 = 1,10,10
data1 = np.random.randint(low, high+1, size)
print("The random numbers for continous:",data)
print("The random numbers for discrete:",data1)
# Create a histogram-like plot
sns.histplot(data, bins=10,kde=True,alpha=0.7, color='blue',label='marks')
sns.histplot(data1, bins=20,alpha=0.8, color='red',label='cgpa')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Uniform Distribution(Continuous and Discrete)')

plt.legend()
plt.show()
```
